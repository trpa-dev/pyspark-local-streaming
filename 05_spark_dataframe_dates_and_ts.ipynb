```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('Basics').getOrCreate()
```

```python
resource_url = '/home/mia/Downloads/Python-and-Spark-for-Big-Data-master/'
df = spark.read.csv(
    resource_url + 'Spark_DataFrames/' + 'appl_stock.csv',
    inferSchema=True,
    header=True
)
```

```python
df.toPandas().head()
```

```python
df.toPandas()[['Date', 'Open']]
```

```python
from pyspark.sql.functions import (
    dayofmonth, hour, dayofyear, month, year, weekofyear,
    format_number, date_format 
)
```

```python
df.select(dayofmonth(df['Date'])).toPandas()
```

```python
df.select(hour(df['Date'])).toPandas()
```

```python
df.select(month(df['Date'])).toPandas()
```

```python
df.select(year(df['Date'])).toPandas()
```

```python
df.withColumn('Year', year(df['Date'])).toPandas()
```

```python
new_df = df.withColumn('Year', year(df['Date']))
```

```python
new_df.toPandas()
```

```python
new_df.groupBy('Year').mean().toPandas()
```

```python
new_df.groupBy('Year').mean() \
    .select(['Year', 'avg(Close)']).toPandas()
```

```python
result = new_df.groupBy('Year').mean().select(['Year', 'avg(Close)'])
```

```python
result.toPandas()
```

```python
new_result = result.withColumnRenamed('avg(Close)', 'Average Closing Price')
```

```python
new_result.toPandas()
```

```python
new_result.select(
    [ 'Year',
      format_number('Average Closing Price', 2).alias('Avg Close')
    ]
).toPandas()
```
